# 04 — CLI: profile a command and parse JSON

This notebook demonstrates:

- running `gpu-profile` as a CLI
- emitting JSON
- parsing the JSON output inside Python

This works even without a GPU; metrics may be `n/a`.


```python
import json, subprocess, shlex

cmd = "gpu-profile --json -- python -c "import time; time.sleep(2)""
print('running:', cmd)

p = subprocess.run(shlex.split(cmd), capture_output=True, text=True)
print('return code:', p.returncode)
print('stdout:', p.stdout.strip())
print('stderr:', p.stderr.strip())

summary = json.loads(p.stdout)
print('keys:', sorted(summary.keys())[:10], '...')
print('util mean:', summary.get('util_gpu_mean'))
```
